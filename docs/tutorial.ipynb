```python
import pandas as pd
import numpy as np
import rose
```

```python
df = pd.read_csv(
    "../sample_data/WBB.csv",
    skiprows=lambda x: x in [0, 1, 2, 3, 4, 5, 7],
)
df.head()
```

```python
df.rose.table(sectors=8)
```

```python
# Plot a standard wind rose
ax = df.rose.plot()
ax.set_title("Windrose for WBB")
```

```python
# Plot an ozone wind rose
# TODO: Find sample data with poor air quality
ax = df.rose.plot(
    var_column="ozone_concentration",
    bins=[0, 60, 75, 95, 115],
    )
ax.set_title("Ozone rose for WBB")
```

```python
df.rose.bar()
```
